```python
import numpy as np
import skimage
import matplotlib.image as mimg
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = mimg.imread('pandas.png')
image.shape
for i in range(image.shape[2]):
    plt.imshow(image[:,:,i])
    plt.show()
```

```python
filter1 = [None] * 6
for i in range(len(filter1)):
    filter1[i] = np.random.rand(5,5,4)
filter1 = np.array(filter1)
filter1.shape
```

```python
def convolution(image,filter_, stride = 1):
    opdim = int((image.shape[1] - filter_.shape[2]) / stride) + 1
    output = np.zeros((opdim, opdim,len(filter_)))
    filterh = filter_.shape[1]
    filterw = filter_.shape[2]
    imageh = image.shape[0]
    imagew = image.shape[1]
    
    for filter_i in range(len(filter_)):
        y=0
        opy = 0
        while y + filterh <= imageh:
            x=0
            opx = 0
            while x + filterw <= imagew:
                output[opy, opx,filter_i] = np.sum(filter_[filter_i] * image[ y:y+filterh, x:x+filterw,:])
                x+=stride
                opx+=1
            y+=stride
            opy += 1
            
    return output
```

```python
output1 = convolution(image, filter1)
print(output1.shape)
```

```python
for i in range(output1.shape[2]):
    plt.imshow(output1[:,:,i])
    plt.show()
```

```python
def maxPool(image, window = 2, stride = 2):
#     print(image.shape)
    outputh = int((image.shape[0] - window) / stride) + 1 
    outputw = int((image.shape[1] - window) / stride) + 1
    outputd= image.shape[2]
    
    output = np.zeros((outputh, outputw,outputd)) 
#     print(output.shape)
    imaged = image.shape[2]
    imageh = image.shape[0]
    imagew = image.shape[1]
    
    for i in range(image.shape[2]):
        y=0
        opy = 0
        while y + window <= imageh:
            x=0
            opx = 0
            while x + window <= imagew:
                output[opy,opx,i] = np.max(image[y:y+window, x:x+window,i])
                x+=stride
                opx+=1
            y += stride
            opy += 1
            
    return output
```

```python
pooledoutput1 = maxPool(output1)
print(pooledoutput1.shape)
```

```python
for i in range(pooledoutput1.shape[2]):
    plt.imshow(pooledoutput1[:,:,i])
    plt.show()
```

```python
filter2 = [None] * 16
for i in range(len(filter2)):
    filter2[i] = np.random.rand(5,5,6)
filter2 = np.array(filter2)
filter2.shape
```

```python
output2 = convolution(pooledoutput1, filter2)
output2.shape
```

```python
pooledoutput2 = maxPool(output2)
print(pooledoutput2.shape)
```

```python
for i in range(pooledoutput2.shape[2]):
    plt.imshow(pooledoutput2[:,:,i])
    plt.show()
```

```python
def softmax(Z):
    E = np.exp(Z - np.max(Z, axis = 1, keepdims=True))
    S = np.sum(E, axis = 1, keepdims= True)
    A = E/S
    return A
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

```python
def fullyconnected(image):
    output = image.reshape((image.shape[0] * image.shape[1] * image.shape[2], 1))
    inputsize = output.shape[0]
    outputsize = 10
    
    numofhiddenlayers = 1
    allweights = [0] * (numofhiddenlayers + 1)
    allz = [0] * (numofhiddenlayers + 1)
    allactivations = [0] * (numofhiddenlayers + 1)
    
    allweights[0] = np.random.randn(100, inputsize)
    allweights[1] = np.random.randn(100, 10)
    
    allz[0] = np.dot(allweights[0], output)
    allactivations[0] = sigmoid(allz[0])
    allz[1] = np.dot(allactivations[0].T, allweights[1])
    allactivations[1] = sigmoid(allz[1])
    
    result = softmax(allactivations[1])
    return result
```

```python
output = fullyconnected(pooledoutput2)
print(np.argmax(output))
```
